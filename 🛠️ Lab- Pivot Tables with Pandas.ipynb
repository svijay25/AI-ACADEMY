## Flattening the DataFrame

We can also **_flatten_** the DataFrame from a multi-hierarchical index to a more traditional one-dimensional index.  We do this by creating each unique combination possible of every level of the multi-hierarchical index.  Since this is a complex task, you do not need to write it -- but take some time to examine the code in the cell below and see if you can understand how it works. 

```python
# Slice the rows where the Population column is equal to 'Not Applicable'
to_drop = data[data['Population'] == 'Not Applicable']

# Drop the offending rows from the DataFrame
data.drop(to_drop.index, axis=0, inplace=True)

# Cast the Population column to an integer data type
data['Population'] = data['Population'].astype('int64')

# Group by 'State' and 'Gender', slice 'Deaths' and 'Population', and apply aggregate functions
aggregated_data = data.groupby(['State', 'Gender'])[['Deaths', 'Population']].agg(['mean', 'min', 'max', 'std'])

# Reset the index of the DataFrame
reset_aggregated_data = aggregated_data.reset_index()

# Flatten the columns
cols0 = reset_aggregated_data.columns.get_level_values(0)
cols1 = reset_aggregated_data.columns.get_level_values(1)
reset_aggregated_data.columns = [
    col0 + "_" + col1 if col1 != "" else col0 for col0, col1 in list(zip(cols0, cols1))
]

# Display the flattened columns
print(reset_aggregated_data.columns)
```

## Stacking and Unstacking DataFrames

Now, let's get some practice stacking and unstacking DataFrames. 

### Stacking

In the cell below, let's display the head of `grouped` to remind ourselves of the format we left it in. 

```python
# Slice the rows where the Population column is equal to 'Not Applicable'
to_drop = data[data['Population'] == 'Not Applicable']

# Drop the offending rows from the DataFrame
data.drop(to_drop.index, axis=0, inplace=True)

# Cast the Population column to an integer data type
data['Population'] = data['Population'].astype('int64')

# Group by 'State' and 'Gender', slice 'Deaths' and 'Population', and apply aggregate functions
grouped = data.groupby(['State', 'Gender'])[['Deaths', 'Population']].agg(['mean', 'min', 'max', 'std'])

# Display the head of the grouped DataFrame
print(grouped.head())
```

As we can see above, `grouped` is currently in a flattened format, with no hierarchical structure to it's indices.  

In the cell below, call the `grouped` DataFrame's `.stack()` method. 

```python
# Slice the rows where the Population column is equal to 'Not Applicable'
to_drop = data[data['Population'] == 'Not Applicable']

# Drop the offending rows from the DataFrame
data.drop(to_drop.index, axis=0, inplace=True)

# Cast the Population column to an integer data type
data['Population'] = data['Population'].astype('int64')

# Group by 'State' and 'Gender', slice 'Deaths' and 'Population', and apply aggregate functions
grouped = data.groupby(['State', 'Gender'])[['Deaths', 'Population']].agg(['mean', 'min', 'max', 'std'])

# Display the head of the grouped DataFrame
print("Head of the grouped DataFrame:")
print(grouped.head())

# Stack the grouped DataFrame
stacked_grouped = grouped.stack()

# Display the head of the stacked DataFrame
print("\nHead of the stacked DataFrame:")
print(stacked_grouped.head())
```

As we can see, the `.stack()` method has stacked our DataFrame from a flattened format into one with a multi-hierarchical index. This is an easy, quick way to aggregate our data.

### Unstacking 

Now, we'll explore unstacking with the `pivot` DataFrame, which is already stacked into a pivot table. 

In the cell below, set unstack `pivot` using the object's `.unstack()` method. Then, display the object to see how it has changed. 

```python
# Pivot the DataFrame
pivot = reset_aggregated_data.pivot(index='State', columns='Gender', values='Deaths_mean')

# Display the head of the pivot DataFrame
print("Head of the pivot DataFrame:")
print(pivot.head())

# Unstack the pivot DataFrame
unstacked_pivot = pivot.unstack()

# Display the unstacked DataFrame
print("\nHead of the unstacked DataFrame:")
print(unstacked_pivot.head())
```

Note that it has unstacked the multi-hierarchical structure of the `pivot` DataFrame by one level. Let's call it one more time and display the results.

In the cell below, set `pivot` equal to `pivot.unstack()` again, and then print the `pivot` object to see how things have changed.

```python
# Slice the rows where the Population column is equal to 'Not Applicable'
to_drop = data[data['Population'] == 'Not Applicable']

# Drop the offending rows from the DataFrame
data.drop(to_drop.index, axis=0, inplace=True)

# Cast the Population column to an integer data type
data['Population'] = data['Population'].astype('int64')

# Group by 'State' and 'Gender', slice 'Deaths' and 'Population', and apply aggregate functions
aggregated_data = data.groupby(['State', 'Gender'])[['Deaths', 'Population']].agg(['mean', 'min', 'max', 'std'])

# Reset the index of the DataFrame
reset_aggregated_data = aggregated_data.reset_index()

# Flatten the columns
cols0 = reset_aggregated_data.columns.get_level_values(0)
cols1 = reset_aggregated_data.columns.get_level_values(1)
reset_aggregated_data.columns = [
    col0 + "_" + col1 if col1 != "" else col0 for col0, col1 in list(zip(cols0, cols1))
]

# Pivot the DataFrame
pivot = reset_aggregated_data.pivot(index='State', columns='Gender', values='Deaths_mean')

# Display the head of the pivot DataFrame
print("Head of the pivot DataFrame:")
print(pivot.head())

# Unstack the pivot DataFrame once
unstacked_pivot = pivot.unstack()

# Display the head of the unstacked DataFrame
print("\nHead of the unstacked DataFrame (after first unstack):")
print(unstacked_pivot.head())

# Unstack the pivot DataFrame again
unstacked_pivot_twice = unstacked_pivot.unstack()

# Display the head of the twice unstacked DataFrame
print("\nHead of the unstacked DataFrame (after second unstack):")
print(unstacked_pivot_twice.head())
```

After calling unstack a second time, we can see that `pivot` has a flattened structure since it has been completely unstacked.
